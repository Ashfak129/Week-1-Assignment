```python
Q3. Explain the use of the following operators using an example:
(i) /
(ii) %
(iii) //
(iv) **
Ans:- (i) / (Division Operator): The division operator / is used to perform regular division between 
two numbers. It returns the quotient as a floating-point number.
Example:
# Regular division
result = 10 / 3
print(result)  # Output: 3.3333333333333335
    (ii) % (Modulo Operator): The modulo operator % is used to find the remainder after the division
of two numbers. It is often used to check divisibility or to wrap values around a certain range.
Example:
# Modulo operator
remainder = 10 % 3
print(remainder)  # Output: 1
    (iii) // (Floor Division Operator): The floor division operator // is used to perform division 
between two numbers and rounds down the result to the nearest integer, discarding any fractional part.
The result is always an integer, even if the numbers involved are floats.
Example:
# Floor division
result = 10 // 3
print(result)  # Output: 3
    (iv) ** (Exponentiation Operator): The exponentiation operator ** is used to raise a number to a 
certain power.
Example:
# Exponentiation
result = 2 ** 3
print(result)  # Output: 8
```

```python
Q7. What do you understand about mutable and immutable data types? Give examples for both showing this property.
```

```python
Ans:- Mutable and immutable are properties of data types in programming languages like Python. These properties
define whether the data stored in a variable of a particular data type can be modified after its creation. Let's
explain both concepts and provide examples for each:
    <i> Mutable Data Types: Mutable data types are those whose values can be changed or modified after creation.
This means that you can update, add, or remove elements from a mutable object without creating a new object. Lists
and dictionaries in Python are examples of mutable data types.
Example of a mutable data type:
# Lists are mutable
my_list = [1, 2, 3, 4]
print("Original List:", my_list)
# Update an element
my_list[2] = 10
print("Modified List:", my_list)
# Add an element
my_list.append(5)
print("List after adding element:", my_list)
# Remove an element
my_list.remove(2)
print("List after removing element:", my_list)

Output:-
Original List: [1, 2, 3, 4]
Modified List: [1, 2, 10, 4]
List after adding element: [1, 2, 10, 4, 5]
List after removing element: [1, 10, 4, 5]

    <ii> Immutable Data Types: Immutable data types are those whose values cannot be changed or modified after
creation. Once an immutable object is created, you cannot change its content. Examples of immutable data types in
Python include integers, floats, strings, and tuples.
Example of an immutable data type:
# Strings are immutable
my_string = "Hello"
print("Original String:", my_string)
# Attempting to update a character (this will raise an error)
# my_string[0] = "h"
# Concatenating to create a new string
new_string = my_string + " World!"
print("New String:", new_string)
# The original string remains unchanged
print("Original String after concatenation:", my_string)

Output:-
Original String: Hello
New String: Hello World!
Original String after concatenation: Hello
```
